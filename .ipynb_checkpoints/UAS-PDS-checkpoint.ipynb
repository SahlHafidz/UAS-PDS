# Visualisasi

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data_ind.csv',low_memory=False);
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
kolom_bentuk = 'bentuk'
propinsi = 'propinsi'

# Baca file CSV


# Filter hanya baris yang memiliki nilai 'sma' di kolom 'bentuk'
df_sd = df[df[kolom_bentuk] == 'SD']

count_sd_per_propinsi = df_sd.groupby(propinsi).size().reset_index(name='Jumlah SD')
# Tampilkan hasil


# Buat diagram lingkaran
plt.figure(figsize=(30, 18))
plt.pie(count_sd_per_propinsi['Jumlah SD'], labels=count_sd_per_propinsi[propinsi], autopct='%1.1f%%', startangle=140)
plt.title('Diagram Lingkaran Jumlah SD per propinsi')
plt.show()
```

# GIS (geographic information system)

```python
import pandas as pd

# Baca file CSV
df = pd.read_csv('data_ind.csv', low_memory=False)

# Set display options
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)

kolom_bentuk = 'bentuk'
propinsi = 'propinsi'


# Filter hanya baris yang memiliki nilai 'sma' di kolom 'bentuk'
df_smk_smp = df[df[kolom_bentuk].isin(['SD','SMP','SMA','SMK','SMLB'])]

# Hitung jumlah SMK dan SMP di setiap kabupaten
count_smk_smp_per_propinsi = df_smk_smp.groupby([propinsi, kolom_bentuk]).size().unstack(fill_value=0).reset_index()

# Rename columns for clarity
count_smk_smp_per_propinsi.columns.name = None
count_smk_smp_per_propinsi = count_smk_smp_per_propinsi.rename(columns={'SD': 'Jumlah SD', 'SMP': 'Jumlah SMP','SMA': 'Jumlah SMA','SMK': 'Jumlah SMK','SMLB': 'Jumlah SMLB'})

# Tampilkan hasil
print(count_smk_smp_per_propinsi)
```

```python
import pandas as pd
from geopy.geocoders import Nominatim
import folium


# Baca file CSV
df = pd.read_csv('data_ind.csv', low_memory=False)
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
kolom_bentuk = 'bentuk'
propinsi = 'propinsi'

# Inisialisasi geolocator
geolocator = Nominatim(user_agent="geo_locator")

# Inisialisasi geolocator dengan timeout yang diperpanjang
geolocator = Nominatim(user_agent="geo_locator", timeout=5)



# Filter hanya baris yang memiliki nilai 'sma' di kolom 'bentuk'
df_sma = df[df[kolom_bentuk].isin(['SD','SMP','SMA','SMK','SMLB'])]

# Hitung jumlah SMA di setiap kabupaten
count_sma_per_propinsi = df_sma.groupby([propinsi, kolom_bentuk]).size().unstack(fill_value=0).reset_index()

count_sma_per_propinsi.columns.name = None
count_sma_per_propinsi = count_sma_per_propinsi.rename(columns={'SD': 'Jumlah SD', 'SMP': 'Jumlah SMP','SMA': 'Jumlah SMA','SMK': 'Jumlah SMK','SMLB': 'Jumlah SMLB'})


# Tambahkan kolom baru 'Latitude' dan 'Longitude'
count_sma_per_propinsi['Latitude'] = 0.0
count_sma_per_propinsi['Longitude'] = 0.0

# Isi data latitude dan longitude berdasarkan nama provinsi
for index, row in count_sma_per_propinsi.iterrows():
    location = geolocator.geocode(row[propinsi] + ', Indonesia')
    if location:
        count_sma_per_propinsi.at[index, 'Latitude'] = location.latitude
        count_sma_per_propinsi.at[index, 'Longitude'] = location.longitude

# Tampilkan hasil
print(count_sma_per_propinsi)

# Buat peta Indonesia
map_indonesia = folium.Map(location=[-2.5489, 118.0149], zoom_start=5)

# Tambahkan marker untuk setiap provinsi

#banten
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Banten':
        # Lokasi manual untuk Provinsi Banten
        count_sma_per_propinsi.at[index, 'Latitude'] = -6.4058
        count_sma_per_propinsi.at[index, 'Longitude'] = 106.0640
#D.I. Yogyakarta
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. D.I. Yogyakarta':
        # Lokasi manual untuk Provinsi D.I. Yogyakarta
        count_sma_per_propinsi.at[index, 'Latitude'] = -7.8012
        count_sma_per_propinsi.at[index, 'Longitude'] = 110.3645
        
#Prov. D.K.I. Jakarta
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. D.K.I. Jakarta':
        # Lokasi manual untuk Provinsi D.K.I. Jakarta
        count_sma_per_propinsi.at[index, 'Latitude'] = -6.2088
        count_sma_per_propinsi.at[index, 'Longitude'] = 106.8456

#Provinsi Gorontalo
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Gorontalo':
        # Lokasi manual untuk Provinsi Gorontalo
        count_sma_per_propinsi.at[index, 'Latitude'] = 0.8292
        count_sma_per_propinsi.at[index, 'Longitude'] = 122.2904

#Prov. Jambi
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Jambi':
        # Lokasi manual untuk Provinsi Jambi
        count_sma_per_propinsi.at[index, 'Latitude'] = -1.4854
        count_sma_per_propinsi.at[index, 'Longitude'] = 102.4389
#Kalimantan Selatan
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Kalimantan Selatan':
        # Lokasi manual untuk Provinsi Kalimantan Selatan
        count_sma_per_propinsi.at[index, 'Latitude'] = -3.3194
        count_sma_per_propinsi.at[index, 'Longitude'] = 114.5901
        
#kalimantan utara
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Kalimantan Utara':
        # Lokasi manual untuk Provinsi Kalimantan Utara
        count_sma_per_propinsi.at[index, 'Latitude'] = 2.4825
        count_sma_per_propinsi.at[index, 'Longitude'] = 117.0182

#Prov. Kalimantan Utara
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Kalimantan Utara':
        # Lokasi manual untuk Provinsi Kalimantan Utara
        count_sma_per_propinsi.at[index, 'Latitude'] = 2.4825
        count_sma_per_propinsi.at[index, 'Longitude'] = 117.0182

#Prov. Kalimantan Utara
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Kalimantan Utara':
        # Lokasi manual untuk Provinsi Kepulauan Riau
        count_sma_per_propinsi.at[index, 'Latitude'] = 0.9160
        count_sma_per_propinsi.at[index, 'Longitude'] = 104.0344
#Prov. Lampung
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Lampung':
        # Lokasi manual untuk Provinsi Lampung
        count_sma_per_propinsi.at[index, 'Latitude'] = -4.5586
        count_sma_per_propinsi.at[index, 'Longitude'] = 105.4068
#Prov. Maluku Utara
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Maluku Utara':
        # Lokasi manual untuk Provinsi Maluku Utara
        count_sma_per_propinsi.at[index, 'Latitude'] = 2.3158
        count_sma_per_propinsi.at[index, 'Longitude'] = 128.3664
        
#Prov. Kepulauan Riau
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Kepulauan Riau':
        # Lokasi manual untuk Provinsi Kepulauan Riau
        count_sma_per_propinsi.at[index, 'Latitude'] = 0.9160
        count_sma_per_propinsi.at[index, 'Longitude'] = 104.0344
# Prov. Sulawesi Barat
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Sulawesi Barat':
        # Lokasi manual untuk Provinsi Sulawesi Barat
        count_sma_per_propinsi.at[index, 'Latitude'] = -2.8518
        count_sma_per_propinsi.at[index, 'Longitude'] = 119.3646
# Prov. Sulawesi Selatan
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Sulawesi Selatan':
        # Lokasi manual untuk Provinsi Sulawesi Selatan
        count_sma_per_propinsi.at[index, 'Latitude'] = -3.3194
        count_sma_per_propinsi.at[index, 'Longitude'] = 119.8907
# Prov. Sulawesi Tengah
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Sulawesi Tengah':
        # Lokasi manual untuk Provinsi Sulawesi Tengah
        count_sma_per_propinsi.at[index, 'Latitude'] = -1.4300
        count_sma_per_propinsi.at[index, 'Longitude'] = 121.4456
# Prov. Sulawesi Utara
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Sulawesi Utara':
        # Lokasi manual untuk Provinsi Sulawesi Utara
        count_sma_per_propinsi.at[index, 'Latitude'] = 0.6247
        count_sma_per_propinsi.at[index, 'Longitude'] = 123.9752
# Prov. Sumatera Selatan
for index, row in count_sma_per_propinsi.iterrows():
    if row[propinsi] == 'Prov. Sumatera Selatan':
        # Lokasi manual untuk Provinsi Sumatera Selatan
        count_sma_per_propinsi.at[index, 'Latitude'] = -3.3194
        count_sma_per_propinsi.at[index, 'Longitude'] = 104.9144
for index, row in count_sma_per_propinsi.iterrows():
    folium.Marker(
        location=[row['Latitude'], row['Longitude']],
        popup=f"{row[propinsi]} - Jumlah SD: {row['Jumlah SD']} - Jumlah SMP: {row['Jumlah SMP']} - Jumlah SMA: {row['Jumlah SMA']} - Jumlah SMK: {row['Jumlah SMK']} - Jumlah SMLB: {row['Jumlah SMLB']}",
        icon=folium.Icon(color='red', icon='info-sign')
    ).add_to(map_indonesia)
# Simpan peta ke file HTML
map_indonesia.save('map_indonesia.html')


# Cetak provinsi yang tidak memiliki koordinat
missing_coordinates = count_sma_per_propinsi[(count_sma_per_propinsi['Latitude'] == 0.0) & (count_sma_per_propinsi['Longitude'] == 0.0)]
print("Provinsi yang tidak memiliki koordinat:")
print(missing_coordinates[[propinsi]])

# Cetak detail hasil geocoding yang tidak berhasil
for index, row in missing_coordinates.iterrows():
    location = geolocator.geocode(row[propinsi] + ', Indonesia')
    if not location:
        print(f"Tidak dapat menemukan koordinat untuk {row[propinsi]}")
```
