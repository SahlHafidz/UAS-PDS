```python
import pandas as pd

df=pd.read_csv('data_ind.csv',low_memory=False);
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
kolom_bentuk = 'bentuk'
propinsi = 'propinsi'

# Baca file CSV


# Filter hanya baris yang memiliki nilai 'sma' di kolom 'bentuk'
df_sma = df[df[kolom_bentuk] == 'SMA']

# Hitung jumlah SMA di setiap kabupaten
count_sma_per_propinsi = df_sma.groupby(propinsi).size().reset_index(name='Jumlah SMA')

# Tampilkan hasil
print(count_sma_per_propinsi)

```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data_ind.csv',low_memory=False);
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
kolom_bentuk = 'bentuk'
propinsi = 'propinsi'

# Baca file CSV


# Filter hanya baris yang memiliki nilai 'sma' di kolom 'bentuk'
df_sd = df[df[kolom_bentuk] == 'SD']

count_sd_per_propinsi = df_sd.groupby(propinsi).size().reset_index(name='Jumlah SD')
# Tampilkan hasil


# Buat diagram lingkaran
plt.figure(figsize=(30, 18))
plt.pie(count_sd_per_propinsi['Jumlah SD'], labels=count_sd_per_propinsi[propinsi], autopct='%1.1f%%', startangle=140)
plt.title('Diagram Lingkaran Jumlah SD per propinsi')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df=pd.read_csv('data_ind.csv',low_memory=False);
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
kolom_bentuk = 'bentuk'
propinsi = 'propinsi'

# Baca file CSV


# Filter hanya baris yang memiliki nilai 'sma' di kolom 'bentuk'
df_sd = df[df[kolom_bentuk] == 'SD']

count_sd_per_propinsi = df_sd.groupby(propinsi).size().reset_index(name='Jumlah SD')
# Tampilkan hasil

# Tampilkan hasil
# print(count_sd_per_kabupaten)

# Buat diagram batang
plt.figure(figsize=(10, 10))
plt.bar(count_sd_per_propinsi[propinsi], count_sd_per_propinsi['Jumlah SD'], color='skyblue')
plt.xlabel('Propinsi')
plt.ylabel('Jumlah SD')
plt.title('Diagram Batang Jumlah SD per Propinsi')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```
