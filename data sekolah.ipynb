```python
import pandas as pd

df=pd.read_csv('data_ind.csv',low_memory=False);
pd.set_option('display.max_rows', None)
pd.set_option('display.max_columns', None)
kolom_bentuk = 'bentuk'
kolom_kabupaten = 'propinsi'

# Baca file CSV


# Filter hanya baris yang memiliki nilai 'sma' di kolom 'bentuk'
df_sma = df[df[kolom_bentuk] == 'SMA']

# Hitung jumlah SMA di setiap kabupaten
count_sma_per_kabupaten = df_sma.groupby(kolom_kabupaten).size().reset_index(name='Jumlah SMA')

# Tampilkan hasil
print(count_sma_per_kabupaten)

```
